```python
def TicTacToe():
    import random
    players = []

    for i in range(1,3):
        print("\n")
        player = input("Enter player's name: ")                         #Read Players' names
        players.append(player)

    play = random.randint(1,2)

    if(play == 1):
        p1 = players[0]
        p2 = players[1]
        print("\n")
        print(p1, "is 'X' and", p2, "is 'O'  (Randomly alloted)" )                          
                                                                     #Randomly allot X and O
    elif(play == 2):
        p1 = players[1]
        p2 = players[0]
        print("\n")
        print(p1, "is 'X' and", p2, "is 'O'  (Randomly alloted)" )   
    

    boardspaces = {'7': ' ' , '8': ' ' , '9': ' ' ,
                   '4': ' ' , '5': ' ' , '6': ' ' ,                      #Board created using Dictionary
                   '1': ' ' , '2': ' ' , '3': ' '  }
            
    def printBoard(board):
        print("\n")
        print(board['7'] + '|' + board['8'] + '|' + board['9'])
        print('-+-+-')
        print(board['4'] + '|' + board['5'] + '|' + board['6'])
        print('-+-+-')
        print(board['1'] + '|' + board['2'] + '|' + board['3'])
    

    turn = 'X'
    chances = 0
    while(True):    
    
                printBoard(boardspaces)
            
                if(turn == 'X'): p = p1
                elif(turn == 'O'): p = p2
            
                print("\n")
                print("It's", p,"'s turn. Your move?: ")

                move = input("Move: ")        

                if boardspaces[move] == ' ':
                    boardspaces[move] = turn
                    chances += 1
                    
                else:
                    print("\n")
                    print("That place is already filled! Choose another space.")
                    continue
                
                if boardspaces['7'] == boardspaces['8'] == boardspaces['9'] != ' ':
                    printBoard(boardspaces)
                    print("\n")
                    print(" ****    And",p,"Wins the game!    ****")                
                    break
                elif boardspaces['4'] == boardspaces['5'] == boardspaces['6'] != ' ':
                    printBoard(boardspaces)              
                    print("\n")
                    print(" ****    And",p,"Wins the game!    ****")                
                    break
                elif boardspaces['1'] == boardspaces['2'] == boardspaces['3'] != ' ':
                    printBoard(boardspaces)              
                    print("\n")
                    print(" ****    And",p,"Wins the game!    ****")                
                    break
                elif boardspaces['1'] == boardspaces['4'] == boardspaces['7'] != ' ':
                    printBoard(boardspaces)                
                    print("\n")
                    print(" ****    And",p,"Wins the game!    ****")                
                    break
                elif boardspaces['2'] == boardspaces['5'] == boardspaces['8'] != ' ':
                    printBoard(boardspaces)               
                    print("\n")
                    print(" ****    And",p,"Wins the game!    ****")                
                    break
                elif boardspaces['3'] == boardspaces['6'] == boardspaces['9'] != ' ':
                    printBoard(boardspaces)                
                    print("\n")
                    print(" ****    And",p,"Wins the game!    ****")                
                    break 
                elif boardspaces['7'] == boardspaces['5'] == boardspaces['3'] != ' ':
                    printBoard(boardspaces)               
                    print("\n")
                    print(" ****    And",p,"Wins the game!    ****")                
                    break
                elif boardspaces['1'] == boardspaces['5'] == boardspaces['9'] != ' ':
                    printBoard(boardspaces)              
                    print("\n")
                    print(" ****    And",p,"Wins the game!    ****")                
                    break    
            
                if turn =='X':
                    turn = 'O'
                                                                                #Change the turn
                else:
                    turn = 'X' 
                
                if chances == 9:               
                    print(" ****    And it's a Tie!    ****")
                    break

TicTacToe()

print("\n")
restart = input("Wanna play Again?(y/n): ")
if restart == "y":
    TicTacToe()
else:
    exit()
```
